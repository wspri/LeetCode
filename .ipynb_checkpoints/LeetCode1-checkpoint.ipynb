##  <center> 练习1 从排序数组中删除重复项 </center>

**给定一个排序数组，你需要在原地删除重复出现的元素，使得每个元素只出现一次，返回移除后数组的新长度。
不要使用额外的数组空间，你必须在原地修改输入数组并在使用 O(1) 额外空间的条件下完成。**

思路：创建一个数组指针，判断指针指向的值与后一个值是否相等，如果相等则将从指针后两个值开始的值全部往前移一位，并删除数组末尾的值。如果不相等，则将指针的值加1。时间复杂度为$O(n^2)$。


``` c++
class Solution {
public:
    int removeDuplicates(vector<int>& nums) 
    {
    if (nums.size()==0)
    return 0;
        int i=0;
        while( i+1<nums.size())
        {
            if (nums[i]==nums[i+1])
            {
                for(int j=i+1;j<nums.size();j++)
                {
                    nums[j]=nums[j+1];
                }
                nums.pop_back();
            }
            else
            {
                i++;
            }
        }
        return nums.size();
    }
};
```

查看阅读解答，官方给出的答案是双指针法，数组完成排序后，可以放置两个指针 $i$ 和 $j$。当$ nums[i] = nums[j]$时，指针$j$加1 跳过重复项。当 $nums[j] \neq nums[i]$时，则将$nums[j]$的值复制到 $nums[i + 1]$。然后递增 $i$，然后继续重复此过程，直到$\ j $到达数组的末尾为止。时间复杂度为$O(n)$。
代码如下：
``` c++
class Solution {
public:
    int removeDuplicates(vector<int>& nums) 
    {
        if (nums.size()==0)
            return 0;
        int i=0;
        for(int j=1;j<nums.size();j++)
        {
            if(nums[i]!=nums[j])
            {
                i++;
                nums[i]=nums[j];
            }
        }
        return i+1;
    }
};
```
运行时间为24ms，但不是最快的。
